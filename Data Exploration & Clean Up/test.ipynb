```python
# create a mock data frame with dates from 2018-01-01 to 2020-12-31 and random values btwn 0 and 100

import pandas as pd
import numpy as np

dates = pd.date_range(start='2018-01-01', end='2018-01-31', freq='D')

data = pd.DataFrame({
    'date_x': dates,
    'var1': np.random.randint(0, 100, len(dates)),
    'var2': np.random.randint(0, 100, len(dates)),
    'var3': np.random.randint(0, 100, len(dates))
})

data
```

```python
# create a new calumn which calculates the moving avaerage of var1 over a 7-day window

data['var1_ma'] = data['var1'].rolling(window=7).mean()

data
```

```python
# create a column which is the percentage change of var1 from the previous day

data['var1_pct_change'] = data['var1'].pct_change()

data
```
